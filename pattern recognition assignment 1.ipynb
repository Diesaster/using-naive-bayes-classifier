## fetching the data then fixing the income column's inconsistencies then putting all the data together to clean both features and targets.

```python
# importing necessary modules and libraries
from ucimlrepo import fetch_ucirepo 
import pandas as pd
adult = fetch_ucirepo(id=2)
# fetching the data.
features = adult.data.features
# the target has "<=50", "<=50.", ">50", ">50." so to edit it to have only 2 targets "<=50" and ">50", we loop and check.
income = pd.DataFrame(["<=50K" if '=' in i else ">50K" for i in adult.data.targets['income']], columns = ['income'])
# we concatenate the features and the income to drop the columns with nulls in both.
features = pd.concat([features, income], axis = 1)
features.head()
```

## checking which columns have nulls.

```python
# finding numbers of nulls in each column.
print(features.isnull().sum())
```

## seeing which data type of columns need to be dealt with.

```python
# noticing that only some categorical columns have nulls.
features.info()
```

## dealing with the nulls by droping the rows.

```python
# droping the rows with nulls.
features.dropna(inplace = True)
print(features)
```

## resetting the indices and removing the new index column.

```python
# resetting indices.
features = features.reset_index()
# droping the index column.
features.drop('index', axis = 1, inplace = True)
print(features)
```

## dropping the duplicates in the data

```python
features.drop_duplicates(inplace = True)
print(features)
```

## resetting the indices and removing the new index column.

```python
# resetting indices.
features = features.reset_index()
# droping the index column.
features.drop('index', axis = 1, inplace = True)
print(features)
```

## separating target from features before one-hot encoding and finding the categorical columns.

```python
# separating the target column from the features before onehotencoding the categorical features.
income = features['income']
features = features.drop('income', axis = 1)
# initializing the categorical columns for the onehotencoding and numeric columns for scaling.
categorical_columns = []
numeric_columns = []
# going throught the feature columns.
for i in range (len(features.columns)):
    # if the data type is "object", add the column to the categorical list.
    if features[features.columns[i]].dtype == 'object':
        categorical_columns.append(features.columns[i])
    else:
        # if not, add the column to the numeric list.
        numeric_columns.append(features.columns[i])
print("categorical columns : ", categorical_columns)
print("numeric columns : ", numeric_columns)
```

## one-hot encoding the categorical features

```python
# importing necessary module.
from sklearn.preprocessing import OneHotEncoder
# making an instance of one-hot encoder.
encoder = OneHotEncoder(drop = 'first')
# transforming the categorical columns using the encoder.
encoded_data = encoder.fit_transform(features[categorical_columns])
# getting the new feature names from the encoder.
feature_names = encoder.get_feature_names_out(categorical_columns)
# making a dataframe from the new features we got.
encoded_df = pd.DataFrame(encoded_data.toarray(), columns = feature_names)
# droping the categorical columns we transformed (non-numeric categorical columns) from the dataset.
features.drop(categorical_columns, axis = 1, inplace = True)
# adding the columns resulted from the transformation to the dataset.
features = pd.concat([features, encoded_df], axis = 1)
features.head()
```

## using feature scaling to standardize the range of the features.

```python
# importing necessary module.
from sklearn.preprocessing import MinMaxScaler
# making an instance of the minmaxscaler.
minmax = MinMaxScaler()
# transforming the numeric columns using the minmaxscaler.
features[numeric_columns] = minmax.fit_transform(features[numeric_columns])
features.head()
```

## splitting the dataset into a training set and a testing set.

```python
# importing necessary module.
from sklearn.model_selection import train_test_split
# making train and testing datasets.
x_train, x_test, y_train, y_test = train_test_split(features, income, test_size = 0.4, random_state = 50)
```

## using naive bayes classifier on the features to predict the income.

```python
# importing necessary module.
from sklearn.naive_bayes import CategoricalNB
# making an instance of the categoricalNB (naive bayes classifier).
naive_bayes = CategoricalNB()
# training the model.
naive_bayes.fit(x_train, y_train)
# predicting using the model.
y_pred = naive_bayes.predict(x_test)
```

## calculating the models performance metrics.

```python
# importing necessary module.
from sklearn.metrics import confusion_matrix, classification_report
# getting the confusion matrix for the actual test and the predicted test.
conf_matrix = confusion_matrix(y_test, y_pred)
print("confusion matrix : ", conf_matrix)
# taking the values of the model metrics
tn, fp, fn, tp = conf_matrix.ravel()
# calculating sensitivity.
sensitivity = tp / (tp + fn) * 100
print("sensitivity : ", sensitivity, "%")
# calculating specificity.
specificity = tn / (tn + fp) * 100
print("specificity : ", specificity, "%")
# finding the classes to see which class is ">50".
print("class order : ", naive_bayes.classes_)
# calculating posterior probability of the features
posterior = naive_bayes.predict_proba(features)
print("posterior probability : ", posterior[:, 1])
# printing classification report
print("classification report : ", classification_report(y_test, y_pred))
```
